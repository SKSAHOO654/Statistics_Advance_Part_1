3 .  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution ?

```python
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p

print(bernoulli_pdf(1, 0.7))
print(bernoulli_pdf(0, 0.7))
```

6 .  Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution ?

```python
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(a, b):
    x = np.arange(a, b + 1)
    cdf = (x - a + 1) / (b - a + 1)

    return x, cdf

def plot_cdf(a, b):
    x, cdf = discrete_uniform_cdf(a, b)

    plt.step(x, cdf, where='post', marker='o', color='blue', label='CDF')
    plt.xlabel('x')
    plt.ylabel('CDF')
    plt.title(f'Cumulative Distribution Function of Discrete Uniform Distribution (a={a}, b={b})')
    plt.grid(True)
    plt.legend()
    plt.show()

def main():
    try:
        a = int(input('Enter the lower bound (a): '))
        b = int(input('Enter the upper bound (b): '))
        if a >= b:
            print('Error: Lower bound (a) must be less than upper bound (b).')
        else:
            plot_cdf(a, b)
    except ValueError:
        print('Invalid input. Please enter integer values.')

if __name__ == '__main__':
    main()
```

9 .  Write a Python function to calculate Z-scores from a dataset and plot them ?

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores from a dataset and plot them.

    Parameters:
    data (list or np.array): Input dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    z_scores = (data - mean) / std_dev

    plt.figure(figsize=(10, 6))
    plt.scatter(range(len(z_scores)), z_scores, color='blue', label='Z-Score')
    plt.axhline(y=0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(y=1, color='green', linestyle='--', label='Z=1')
    plt.axhline(y=-1, color='green', linestyle='--', label='Z=-1')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-Score')
    plt.title('Z-Score Plot')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

data = [50, 55, 60, 45, 70, 80, 65, 85, 90, 40]
z_scores = calculate_and_plot_zscores(data)
print("Z-Scores:", z_scores)
```

10 .  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution ?

```python
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem_demo(sample_size=30, num_samples=1000):
    population = np.random.uniform(0, 10, 100000)

    sample_means = []

    for _ in range(num_samples):
        sample = np.random.choice(population, sample_size)
        sample_means.append(np.mean(sample))

    plt.figure(figsize=(10, 6))
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
    plt.title(f'Central Limit Theorem Demonstration (n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

central_limit_theorem_demo(sample_size=5)
central_limit_theorem_demo(sample_size=30)
central_limit_theorem_demo(sample_size=100)
```

18 .  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution ?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (data_point - mean) / std_dev
    return z_score, mean, std_dev

def plot_standard_normal_distribution(z_score):

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, color='blue', label='Standard Normal Distribution')
    plt.fill_between(x, y, color='blue', alpha=0.2)

    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-Score = {z_score:.2f}')
    plt.axvline(0, color='green', linestyle='--', label='Mean ( = 0)')

    plt.title('Standard Normal Distribution with Z-Score')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()

data = np.random.normal(50, 10, 1000)
data_point = 65

z_score, mean, std_dev = calculate_z_score(data_point, data)
print(f"Data Point: {data_point}")
print(f"Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}")
print(f"Calculated Z-Score: {z_score:.2f}")

plot_standard_normal_distribution(z_score)
```

19 .  Implement hypothesis testing using Z-statistics for a sample dataset ?

```python
import numpy as np
from scipy.stats import norm

def hypothesis_testing(data, population_mean, population_std_dev, alpha=0.05):
    sample_mean = np.mean(data)
    sample_size = len(data)

    z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    critical_value = norm.ppf(1 - alpha / 2)

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-Score: {z_score:.2f}")
    print(f"Critical Value (={alpha}): {critical_value:.2f}")

    if abs(z_score) > critical_value:
        print("Reject the Null Hypothesis (H0). Significant difference detected.")
    else:
        print("Fail to Reject the Null Hypothesis (H0). No significant difference detected.")

np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=50)

population_mean = 50
population_std_dev = 10
alpha = 0.05

hypothesis_testing(sample_data, population_mean, population_std_dev, alpha)
```

20 . Create a confidence interval for a dataset using Python and interpret the result ?

```python
import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(data, confidence_level=0.95):
    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1)
    sample_size = len(data)

    alpha = 1 - confidence_level
    z_value = norm.ppf(1 - alpha / 2)

    margin_of_error = z_value * (sample_std_dev / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Standard Deviation: {sample_std_dev:.2f}")
    print(f"Z-Value for {confidence_level*100}% Confidence Level: {z_value:.2f}")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)

calculate_confidence_interval(sample_data, confidence_level=0.95)
```

21 . Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean ?

```python
import numpy as np
from scipy.stats import norm

def generate_data_and_calculate_ci(mean=50, std_dev=10, sample_size=100, confidence_level=0.95):
    data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1)

    alpha = 1 - confidence_level
    z_value = norm.ppf(1 - alpha / 2)

    margin_of_error = z_value * (sample_std_dev / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
    print(f"Z-Value for {confidence_level*100}% Confidence Level: {z_value:.2f}")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    print("\nInterpretation:")
    print(f"We are {confidence_level*100:.0f}% confident that the true population mean lies within the interval "
          f"({lower_bound:.2f}, {upper_bound:.2f}).")

generate_data_and_calculate_ci(mean=50, std_dev=10, sample_size=100, confidence_level=0.95)
```

22 .  Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution ?

23 . Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution ?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam=5, k_max=20):
    k_values = np.arange(0, k_max + 1)

    cdf_values = poisson.cdf(k_values, lam)

    plt.figure(figsize=(10, 6))
    plt.step(k_values, cdf_values, where='mid', color='blue', label=f'Poisson CDF (={lam})')

    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution with ={lam}')
    plt.xlabel('Number of Events (k)')
    plt.ylabel('Cumulative Probability')
    plt.grid(True)
    plt.legend(loc='lower right')
    plt.show()

    for k, cdf in zip(k_values, cdf_values):
        print(f"P(X  {k}) = {cdf:.4f}")

plot_poisson_cdf(lam=5, k_max=20)
```

24 . Simulate a random variable using a continuous uniform distribution and calculate its expected value ?

```python
import numpy as np

def simulate_continuous_uniform(a=0, b=10, size=10000):
    random_variables = np.random.uniform(a, b, size)

    sample_mean = np.mean(random_variables)

    theoretical_mean = (a + b) / 2

    print(f"Generated {size} Random Variables from Uniform({a}, {b})")
    print(f"Theoretical Expected Value: {theoretical_mean}")
    print(f"Sample Mean (Empirical Expected Value): {sample_mean:.4f}")

    return random_variables, theoretical_mean, sample_mean

simulate_continuous_uniform(a=5, b=15, size=10000)
```

25 . Write a Python program to compare the standard deviations of two datasets and visualize the difference ?

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(mean1=50, std_dev1=10, mean2=50, std_dev2=20, size=1000):
    data1 = np.random.normal(mean1, std_dev1, size)
    data2 = np.random.normal(mean2, std_dev2, size)

    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    print(f"Dataset 1 - Mean: {np.mean(data1):.2f}, Standard Deviation: {std1:.2f}")
    print(f"Dataset 2 - Mean: {np.mean(data2):.2f}, Standard Deviation: {std2:.2f}")

    plt.figure(figsize=(10, 6))
    plt.hist(data1, bins=30, alpha=0.5, color='blue', label=f'Dataset 1 (std = {std1:.2f})')
    plt.hist(data2, bins=30, alpha=0.5, color='orange', label=f'Dataset 2 (std = {std2:.2f})')

    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Comparison of Standard Deviations')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()

compare_standard_deviations(mean1=50, std_dev1=10, mean2=50, std_dev2=20)
```

26 .  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution ?

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_range_and_iqr(mean=50, std_dev=10, size=1000):
    data = np.random.normal(mean, std_dev, size)

    data_range = np.max(data) - np.min(data)

    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    print(f"Dataset Summary:")
    print(f"Mean: {np.mean(data):.2f}, Standard Deviation: {np.std(data, ddof=1):.2f}")
    print(f"Range: {data_range:.2f}")
    print(f"Q1 (25th Percentile): {q1:.2f}")
    print(f"Q3 (75th Percentile): {q3:.2f}")
    print(f"Interquartile Range (IQR): {iqr:.2f}")

    plt.figure(figsize=(8, 5))
    plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
    plt.title('Box Plot of Normally Distributed Data')
    plt.xlabel('Values')
    plt.grid(True)
    plt.show()

calculate_range_and_iqr(mean=50, std_dev=10, size=1000)
```

27 . Implement Z-score normalization on a dataset and visualize its transformation ?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores, mean, std_dev

def visualize_normalization(original_data):
    normalized_data, mean, std_dev = z_score_normalization(original_data)

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=30, color='blue', alpha=0.6)
    plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
    plt.axvline(mean + std_dev, color='green', linestyle='--', label=f"1 Std Dev = {std_dev:.2f}")
    plt.axvline(mean - std_dev, color='green', linestyle='--')
    plt.title('Original Data')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=30, color='orange', alpha=0.6)
    plt.axvline(0, color='red', linestyle='--', label="Mean = 0")
    plt.axvline(1, color='green', linestyle='--', label="1 Std Dev = 1")
    plt.axvline(-1, color='green', linestyle='--')
    plt.title('Z-Score Normalized Data')
    plt.xlabel('Z-Scores')
    plt.ylabel('Frequency')
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f"Original Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}")
    print(f"After Normalization -> Mean: {np.mean(normalized_data):.2f}, Standard Deviation: {np.std(normalized_data):.2f}")

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

visualize_normalization(data)
```

28 .  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution ?

```python
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

def calculate_skewness_and_kurtosis(mean=0, std_dev=1, size=1000):
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    data_skewness = skew(data)
    data_kurtosis = kurtosis(data, fisher=True)

    print(f"Dataset Summary:")
    print(f"Mean: {np.mean(data):.2f}, Standard Deviation: {np.std(data, ddof=1):.2f}")
    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (Excess Kurtosis): {data_kurtosis:.4f}")

    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Normally Distributed Data')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

calculate_skewness_and_kurtosis(mean=0, std_dev=1, size=1000)
```
